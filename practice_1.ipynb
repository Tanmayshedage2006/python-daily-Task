```python
num = int(input("Enter a number: ")) 
fact = 1 
if num < 0: 
    print("Factorial of number is not possible") 
elif num == 0: 
    print("Factorial of 0 is 1")
else:                                         # factorial number using if for 
    for i in range(1, num + 1): 
        fact= fact * i 
        print("Factorial of", num, "is", fact)
```

```python
number = list(range(1, 10)) 
odd = [] 
even = [] 
for num in number:
    if num % 2 == 0: 
        even.append(num) 
    else:                              # Even odd number using list and it's operation
        odd.append(num) 
        print("Odd numbers:", odd)
        print("Even numbers:", even)
```

```python
numbers = [10, 25, 5, 80, 45] 
print("Maximum number:", max(numbers))     # find out min and max using max, min function
print("Minimum number:", min(numbers))
```

```python
data = [10, 20, 30, 40, 50, 60] 
print("Original List:", data) 
print("First three elements:", data[:3]) 
print("Last three elements:", data[-3:])           # Slicing of a list 
print("Middle elements:", data[2:5]) 
```

```python
set1 = {1, 2, 3, 4} 
set2 = {3, 4, 5, 6} 
                                                 # set and perform a operation on it
print("Union:", set1 | set2) 
print("Intersection:", set1 & set2) 
print("Difference:", set1 - set2)
```
