```python
def sum_numbers(a, b):
    total = a + b
    print("Sum =", total)   # Prints the sum
    return total            # Returns the sum

# Function Call
result = sum_numbers(10, 20)

print("Returned Value =", result)
```

```python
# Global variable
counter = 0

# Function
def increment():
    global counter
    counter = counter + 1
    print("Counter =", counter)

# Function Calls
increment()
increment()
increment()
```

```python
# Function
def demo_local():
    message = "Hello Local"   # Local variable
    print(message)

# Function Call
demo_local()

# Trying to access the local variable outside the function
print(message)
```

```python
# Global variable
message = "Hello, Global"

# Function
def demo_local():
    print(message)   # Accessing the global variable

# Function Call
demo_local()
```

```python
# Function
def greet(name):
    """
    Purpose:
        Returns a greeting message.

    Parameter:
        name (str): The name of the person.

    Returns:
        str: A greeting message.
    """
    return "Hello, " + name + "!"

# Function Call
message = greet("John")
print(message)

# Display the docstring
help(greet)
```

```python
# Function
def calculate_area(radius):
    """
    Purpose:
        Calculates the area of a circle.

    Formula:
        Area = 3.14 × radius × radius

    Parameter:
        radius (float): The radius of the circle.

    Returns:
        float: The area of the circle.
    """
    return 3.14 * radius * radius

# Function Call
area = calculate_area(5)

print("Area of Circle =", area)

# Display the docstring
help(calculate_area)
```
