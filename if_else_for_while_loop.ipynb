# While loop Question

```python
num = int(input("Enter a number: "))

while num >= 0:   # it's reverse order function
    print(num)
    num -= 1
```

```python
num = 3

while num != 5:
    print(num)
    num = int(input("Enter 5 to stop the loop: "))
```

```python
sum = 0

num = int(input("Enter a number: "))

while num != 0:
    sum = sum + num
    num = int(input("Enter a number: "))

print("Total Sum =", sum)
```

```python
num = int(input("Enter a number: "))

reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reversed Number =", reverse)
```
